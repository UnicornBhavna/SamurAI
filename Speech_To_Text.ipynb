```python
# Step 3: Import Libraries
from faster_whisper import WhisperModel
import os
import subprocess
import librosa
import numpy as np
```

```python
# Step 4: Load the WhisperX Model
# Choose the model size based on your needs: tiny, base, small, medium, large
import torch

model_size = "base"  # Options: tiny, base, small, medium, large
device = "cuda" if torch.cuda.is_available() else "cpu"  # Use GPU if available
compute_type = "float16" if device == "cuda" else "int8"  # Use float16 for GPU, int8 for CPU
```

```python
# Load the WhisperX model
model = WhisperModel(model_size, device=device, compute_type=compute_type)
```

```python
# Step 5: Define a Function to Extract Audio from Video
def extract_audio_from_video(video_path, output_audio_path):
    # Use ffmpeg to extract audio from video
    command = f"ffmpeg -i {video_path} -q:a 0 -map a {output_audio_path} -y"
    subprocess.call(command, shell=True)
```

```python
# Step 6: Define a Function to Transcribe Audio
def transcribe_audio(audio_path):
    # Load audio using librosa (resample to 16kHz if necessary)
    audio, _ = librosa.load(audio_path, sr=16000)

    # Transcribe audio using WhisperX
    segments, info = model.transcribe(audio)

    # Combine all segments into a single transcription
    transcription = " ".join([segment.text for segment in segments])
    return transcription
```

```python
# Step 7: Process All Videos in a Google Drive Folder
def process_videos_in_folder(folder_path, audio_output_folder, text_output_folder):
    # Create the output folders if they don't exist
    os.makedirs(audio_output_folder, exist_ok=True)
    os.makedirs(text_output_folder, exist_ok=True)

    # Iterate through all files in the folder
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)

        # Check if the file is a video (MP4)
        if file_name.endswith(".mp4"):
            print(f"Processing video: {file_name}")

            # Extract audio from the video
            output_audio_path = os.path.join(audio_output_folder, file_name.replace(".mp4", ".wav"))
            extract_audio_from_video(file_path, output_audio_path)

            # Transcribe the audio
            transcription = transcribe_audio(output_audio_path)

            # Save the transcription to a text file
            output_text_path = os.path.join(text_output_folder, file_name.replace(".mp4", ".txt"))
            with open(output_text_path, "w") as f:
                f.write(transcription)

            print(f"Audio saved to: {output_audio_path}")
            print(f"Transcription saved to: {output_text_path}")
```

```python
# Step 8: Specify Input and Output Folders
# Replace with your Google Drive folder path containing videos
input_folder = "/content/drive/MyDrive/Dementia/Videos"  # Replace with your folder path
audio_output_folder = "/content/drive/MyDrive/Dementia/Voice"  # Folder to save extracted audio
text_output_folder = "/content/drive/MyDrive/Dementia/Transcriptions"  # Folder to save transcriptions
```
